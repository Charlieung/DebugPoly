```python
def url_to_soup(url):
    response = requests.get(url)
    return BeautifulSoup(response.content, "html.parser")
```

```python
def extract_actual_article_url(article_soup):
    return article_soup.find_all('div', 'article-link-hidden')[0].get_text().strip()
```

```python
from newspaper import Article

def extract_article(url):
    print(url)
    soup = url_to_soup(url)
    bias_div = soup.find(id=BIAS_DIV_ID)
    bias_text = bias_div.img["title"]
    source = bias_div.a.get_text()
    article_url = extract_actual_article_url(soup)
    article = Article(article_url, language='en')
    article.download()
    article.parse()
    article.nlp()
    return {
        "title": article.title,
        "text": article.text,
        "bias": bias_text,
        "source": source,
        "authors": article.authors,
        "keywords": article.keywords
    }
```

```python
def urls_to_df(urls):
    extracted = [extract_article(url) for url in urls]
    return pd.DataFrame(extracted, index=range(len(extracted)))
```

```python
import extract_article
import pandas as pd
from tqdm import tqdm

urls = pd.read_csv('article_urls.csv', index_col="index", names=["index", "urls"])

STEP_SIZE = 10
n_rows = urls["urls"].shape[0]
for ii in tqdm(range(0, n_rows, STEP_SIZE)):
    slice_end = min(ii+STEP_SIZE, n_rows)
    slice = urls["urls"].iloc[ii:slice_end]
    slice_frame = extract_article.urls_to_df(slice)
    filename = "dataset/article_data_slice" + str(ii) + "-" + str(slice_end) + ".csv"
    slice_frame.to_csv(filename)
    print(ii, min(ii+STEP_SIZE, n_rows))
```
